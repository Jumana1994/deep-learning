```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
```

```python
# data=pd.DataFrame(data["quality"],colomns=[1,2,3,4,11])
```

```python
data['fixed acidity']=scale.fit_transform(data[['fixed acidity']])
data['volatile acidity']=scale.fit_transform(data[['volatile acidity']])
data['citric acid']=scale.fit_transform(data[['citric acid']])
data['residual sugar']=scale.fit_transform(data[['residual sugar']])
data['alcohol']=scale.fit_transform(data[['alcohol']])
data['chlorides']=scale.fit_transform(data[['chlorides']])
data['free sulfur dioxide']=scale.fit_transform(data[['free sulfur dioxide']])
data['total sulfur dioxide']=scale.fit_transform(data[['total sulfur dioxide']])
data['sulphates']=scale.fit_transform(data[['sulphates']])
data['pH']=scale.fit_transform(data[['pH']])
data['density']=scale.fit_transform(data[['density']])
data['quality']=scale.fit_transform(data[['quality']])
```

```python
data
```

```python
x=data[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'alcohol', 'quality','chlorides', 'free sulfur dioxide','total sulfur dioxide', 'density',
       'pH', 'sulphates']]
```

```python
k_range=range(1,11)
from sklearn.cluster import KMeans
result=[]
for i in k_range:
    model=KMeans(n_clusters=i)
    model.fit_predict(x)
    result.append(model.inertia_)
```

```python
result
```

```python
plt.plot(k_range,result,marker='o',color='b')
plt.xlabel("clusters")
plt.ylabel("values")
```

```python
model=KMeans(n_clusters=3)
y_pred=model.fit_predict(x)
y_pred
```

```python
data['group']=y_pred
```
